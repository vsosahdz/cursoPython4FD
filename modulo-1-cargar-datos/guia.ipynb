## 1. ¿Cómo funciona un notebook?

Un notebook se compone de **celdas**. Hay dos tipos:

- **Celdas de código**, con instrucciones de Python que puedes ejecutar.

Para **ejecutar** una celda de código: haz clic en ella y presiona
**`Shift + Enter`**. El resultado aparece justo debajo.

Prueba con la siguiente celda 👇

```python
# Todo lo que empieza con # es un COMENTARIO: Python lo ignora, es para ti.
# Esta es una celda de código. Presiona Shift+Enter para ejecutarla.
2 + 2
```

## 2. Cargar los datos

Usaremos la librería **pandas**, la herramienta estándar para manejar tablas de
datos en Python. La "importamos" y por costumbre la llamamos `pd`.

```python
import pandas as pd  # 'pd' es el apodo con el que usaremos pandas
```

Ahora cargamos el archivo `ventas.csv` con `read_csv` (leer CSV).
Un CSV es un archivo de tabla, como un Excel pero en texto plano.

Guardamos la tabla en una variable llamada `ventas`.

```python
ventas = pd.read_csv("../datos/ventas.csv")
```

## 3. Ver los datos

### `head()` — los primeros registros
Como asomarte a las primeras filas de tu hoja de cálculo.

```python
ventas.head()   # por defecto muestra las primeras 5 filas
```

```python
ventas.head(3)  # también puedes pedir un número específico
```

### `tail()` — los últimos registros

```python
ventas.tail()
```

### `shape` — ¿cuántas filas y columnas tiene?
Devuelve `(número_de_filas, número_de_columnas)`.

```python
ventas.shape
```

```python
# Lo podemos leer de forma más clara:
filas, columnas = ventas.shape
print("La tabla tiene", filas, "filas y", columnas, "columnas")
```

### `columns` — ¿cómo se llaman las columnas?

```python
ventas.columns
```

### `info()` — un resumen de la tabla
Muestra cada columna, cuántos valores tiene y de qué **tipo** es
(número, texto, etc.). Los tipos los veremos a fondo en el Módulo 2.

```python
ventas.info()
```

## 4. Seleccionar una columna

Para tomar una sola columna, escribimos su nombre entre corchetes y comillas.
Es como seleccionar la columna "monto_venta" completa en Excel.

```python
ventas["monto_venta"].head()
```

También podemos cargar la segunda tabla del taller, `concesionarios.csv`,

## ✅ Resumen del módulo

| Quiero... | En Excel | En pandas |
|-----------|----------|-----------|
| Abrir una tabla | Abrir el archivo | `pd.read_csv("archivo.csv")` |
| Ver las primeras filas | Mirar arriba | `.head()` |
| Ver cuántas filas hay | Ir al final | `.shape` |
| Ver los nombres de columnas | Fila de encabezados | `.columns` |
| Un resumen de la tabla | — | `.info()` |
| Seleccionar una columna | Clic en el encabezado | `tabla["columna"]` |
