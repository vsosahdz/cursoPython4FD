No es lo mismo un **número** (puedes sumarlo, promediarlo) que un **texto**
(una categoría) o una **fecha**. Si Python confunde los tipos, tus análisis
saldrán mal.

En este módulo aprenderás a:
- Ver de qué tipo es cada columna.
- Distinguir datos **categóricos** de **numéricos** y de **fecha**.
- Convertir una columna al tipo correcto.

> 🌉 **Puente con Excel:** es como el "Formato de celda" en Excel
> (Texto / Número / Fecha). Si una fecha quedó como texto, no puedes ordenarla
> ni restar fechas correctamente.

```python
import pandas as pd
ventas = pd.read_csv("../datos/ventas.csv")
ventas.head(3)
```

## 1. ¿De qué tipo es cada columna?

`dtypes` nos dice el tipo de cada columna:
- `int64` / `float64` → **números** (enteros / con decimales)
- `object` → normalmente **texto** (categorías)

```python
ventas.dtypes
```

## 2. Datos categóricos

Un dato **categórico** es una etiqueta que se repite y agrupa: el `modelo`,
la `categoria_vehiculo`, el `metodo_pago`. No tiene sentido "promediarlos".

### `unique()` — ¿qué valores distintos existen?

```python
ventas["modelo"].unique()
```

### `value_counts()` — ¿cuántas veces aparece cada valor?
Como una mini tabla dinámica de conteo.

```python
ventas["metodo_pago"].value_counts()
```

### Convertir a tipo `category`
Marcar una columna como categórica ahorra memoria y deja claro que es una etiqueta.

```python
ventas["modelo"] = ventas["modelo"].astype("category")
ventas["modelo"].dtype
```

## 3. Datos numéricos

Con los números **sí** tienen sentido las operaciones: sumar, promediar, etc.

```python
print("Monto promedio de venta:", ventas["monto_venta"].mean())
print("Monto máximo:", ventas["monto_venta"].max())
print("Total de unidades vendidas:", ventas["unidades"].sum())
```

### `describe()` — estadística rápida de las columnas numéricas
Cuenta, promedio, mínimo, máximo y percentiles de un solo golpe.

```python
ventas.describe()
```

## 4. Fechas

La columna `fecha` se cargó como **texto** (`object`). Para poder trabajarla
como fecha (ordenar, sacar el mes, restar días) hay que **convertirla**.

```python
# Antes: es texto
print("Tipo ANTES:", ventas["fecha"].dtype)

ventas["fecha"] = pd.to_datetime(ventas["fecha"])

print("Tipo DESPUÉS:", ventas["fecha"].dtype)
```

Ya como fecha, podemos extraer partes como el **año** o el **mes**:

```python
ventas["anio"] = ventas["fecha"].dt.year
ventas["mes"] = ventas["fecha"].dt.month
ventas[["fecha", "anio", "mes"]].head()
```

## ✅ Resumen del módulo

| Tipo | Ejemplos en el dataset | Qué puedes hacer |
|------|------------------------|------------------|
| **Categórico** (texto) | `modelo`, `metodo_pago` | contar, agrupar |
| **Numérico** | `monto_venta`, `unidades` | sumar, promediar |
| **Fecha** | `fecha` | ordenar, sacar mes/año |

| Quiero... | En pandas |
|-----------|-----------|
| Ver tipos de cada columna | `.dtypes` |

| Conteo por categoría | `columna.value_counts()` |
| Estadística rápida | `.describe()` |
| Texto → fecha | `pd.to_datetime(columna)` |
