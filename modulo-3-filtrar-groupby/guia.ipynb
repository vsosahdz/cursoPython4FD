Aquí está el corazón del análisis de datos:
- **Filtrar**: quedarte solo con las filas que te interesan (autofiltro de Excel).
- **Agrupar** (`groupby`): resumir por categoría (¡la tabla dinámica!).
- **Combinar tablas** (`merge`): traer datos de otra tabla (BUSCARV / VLOOKUP).

> 🌉 **Puente con Excel:** filtrar = Autofiltro · agrupar = Tabla dinámica ·
> combinar = BUSCARV.

```python
import pandas as pd
ventas = pd.read_csv("../datos/ventas.csv")
concesionarios = pd.read_csv("../datos/concesionarios.csv")
ventas["fecha"] = pd.to_datetime(ventas["fecha"])
ventas.head(3)
```

## 1. Filtrar filas (condiciones)

Escribimos una **condición** dentro de los corchetes. pandas devuelve solo las
filas donde la condición es verdadera.

Ejemplo: solo las ventas del modelo **Mustang**.

```python
mustang = ventas[ventas["modelo"] == "Mustang"]
mustang.head()
```

```python
# ¿Cuántas ventas de Mustang hubo?
print("Ventas de Mustang:", mustang.shape[0])
```

### Condiciones numéricas
Ventas con monto mayor a 1 millón:

```python
grandes = ventas[ventas["monto_venta"] > 1_000_000]
grandes.shape[0]
```

### Varias condiciones a la vez
Usa `&` para "Y", `|` para "O". Cada condición va entre paréntesis.

Ventas de Mustang **Y** pagadas de Contado:

```python
filtro = ventas[(ventas["modelo"] == "Mustang") & (ventas["metodo_pago"] == "Contado")]
filtro.head()
```

## 2. `loc` e `iloc` — seleccionar filas y columnas

- **`loc`**: selecciona por **etiqueta** (nombre de columna / condición).
- **`iloc`**: selecciona por **posición** (número de fila/columna).

```python
# loc: filas donde categoria es Pickup, mostrando solo algunas columnas
ventas.loc[ventas["categoria_vehiculo"] == "Pickup", ["modelo", "monto_venta"]].head()
```

```python
# iloc: las primeras 3 filas y primeras 4 columnas (por posición)
ventas.iloc[0:3, 0:4]
```

## 3. Agrupar con `groupby` (la tabla dinámica)

`groupby` resume los datos por categoría. La receta es:

`tabla.groupby("columna_categoria")["columna_a_resumir"].operacion()`

Total de ventas por modelo:

```python
ventas.groupby("modelo")["monto_venta"].sum()
```

Promedio de monto por método de pago:

```python
ventas.groupby("metodo_pago")["monto_venta"].mean()
```

### Varias operaciones a la vez con `agg`
Por modelo: cuántas ventas, total y promedio.

```python
ventas.groupby("modelo")["monto_venta"].agg(["count", "sum", "mean"])
```

## 4. Combinar tablas con `merge` (BUSCARV)

La tabla `ventas` tiene `id_concesionario`, pero no la **región**. La región
está en `concesionarios`. Los unimos por la columna en común, igual que un
BUSCARV trae datos de otra hoja.

```python
ventas_completa = ventas.merge(concesionarios, on="id_concesionario")
ventas_completa[["modelo", "monto_venta", "region", "ciudad"]].head()
```

Ahora sí podemos agrupar por **región** (que vino de la otra tabla):

```python
ventas_completa.groupby("region")["monto_venta"].sum().sort_values(ascending=False)
```

## ✅ Resumen del módulo

| Quiero... | En Excel | En pandas |
|-----------|----------|-----------|
| Quedarme con ciertas filas | Autofiltro | `tabla[tabla["col"] == valor]` |
| Filas por posición | — | `.iloc[filas, columnas]` |
| Filas/columnas por nombre | — | `.loc[condición, ["col"]]` |
| Resumir por categoría | Tabla dinámica | `.groupby("col")["otra"].sum()` |
| Traer datos de otra tabla | BUSCARV | `.merge(otra, on="clave")` |
