# Módulo 5 — Visualización con matplotlib 📈

**Taller Python for Ford 2026**

Una gráfica comunica más que una tabla de números. Aprenderás a crear los
3 gráficos más útiles en finanzas:
- **Línea**: una tendencia en el tiempo (ventas por mes).
- **Barras**: comparar categorías (ventas por región).
- **Histograma**: ver la distribución de un valor (montos de venta).

> 🌉 **Puente con Excel:** es lo mismo que Insertar → Gráfico, pero escrito.

```python
import pandas as pd
import matplotlib.pyplot as plt

ventas = pd.read_csv("../datos/ventas.csv")
concesionarios = pd.read_csv("../datos/concesionarios.csv")
ventas["fecha"] = pd.to_datetime(ventas["fecha"])
ventas = ventas.merge(concesionarios, on="id_concesionario")
```

## 1. Gráfico de línea — ventas por mes

Primero resumimos las ventas por mes (con lo que aprendimos de agrupar),
y luego lo graficamos.

```python
# Agrupar por mes: creamos una columna 'mes' (primer día de cada mes)
ventas["mes"] = ventas["fecha"].dt.to_period("M").dt.to_timestamp()
ventas_por_mes = ventas.groupby("mes")["monto_venta"].sum()

plt.figure(figsize=(10, 4))
plt.plot(ventas_por_mes.index, ventas_por_mes.values, marker="o")
plt.title("Ventas totales por mes")
plt.xlabel("Mes")
plt.ylabel("Monto de ventas (MXN)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 2. Gráfico de barras — ventas por región

Ideal para comparar categorías.

```python
ventas_por_region = ventas.groupby("region")["monto_venta"].sum().sort_values()

plt.figure(figsize=(8, 4))
plt.bar(ventas_por_region.index, ventas_por_region.values, color="#1f5fa8")
plt.title("Ventas totales por región")
plt.xlabel("Región")
plt.ylabel("Monto de ventas (MXN)")
plt.tight_layout()
plt.show()
```

### Barras horizontales (útil con muchas categorías o nombres largos)
Ventas por modelo:

```python
ventas_por_modelo = ventas.groupby("modelo")["monto_venta"].sum().sort_values()

plt.figure(figsize=(8, 4))
plt.barh(ventas_por_modelo.index, ventas_por_modelo.values, color="#2e8b57")
plt.title("Ventas totales por modelo")
plt.xlabel("Monto de ventas (MXN)")
plt.tight_layout()
plt.show()
```

## 3. Histograma — distribución de los montos

Un histograma muestra **cómo se reparten** los valores: ¿la mayoría de las
ventas son chicas o grandes? Agrupa los montos en "cajones" y cuenta cuántas
ventas caen en cada uno.

```python
plt.figure(figsize=(9, 4))
plt.hist(ventas["monto_venta"], bins=30, color="#c1440e", edgecolor="white")
plt.title("Distribución de los montos de venta")
plt.xlabel("Monto de venta (MXN)")
plt.ylabel("Número de ventas")
plt.tight_layout()
plt.show()
```

## 4. Atajo: graficar directo desde pandas

pandas tiene un `.plot()` que usa matplotlib por debajo. Rápido para explorar.

```python
ventas.groupby("categoria_vehiculo")["monto_venta"].sum().plot(
    kind="bar", title="Ventas por categoría de vehículo", figsize=(8, 4)
)
plt.tight_layout()
plt.show()
```

## ✅ Resumen del módulo

| Quiero mostrar... | Gráfico | Comando |
|-------------------|---------|---------|
| Tendencia en el tiempo | Línea | `plt.plot(x, y)` |
| Comparar categorías | Barras | `plt.bar(x, y)` / `plt.barh()` |
| Distribución de un valor | Histograma | `plt.hist(datos, bins=30)` |
| Explorar rápido | — | `tabla.plot(kind="bar")` |

Elementos clave de toda gráfica: `title` (título), `xlabel`/`ylabel` (ejes).
